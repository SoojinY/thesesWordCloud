# KCI 국내학술지 인용색인 정보 포탈

https://www.kci.go.kr/kciportal/po/openapi/openApiKeyRequestView.kci

Key : 	88687611

```python
# 라이브러리 import
import requests
import pprint

# url 입력
# keyword = '코로나'
url = 'https://open.kci.go.kr/po/openapi/openApiSearch.kci?apiCode=articleSearch&key=88687611&keyword=%EC%BD%94%EB%A1%9C%EB%82%98&displayCount=50'

# url 불러오기
response = requests.get(url, verify=False)

#데이터 값 출력해보기
contents = response.text
```

# XML -> DataFame 변환 

https://wonhwa.tistory.com/16

```python
# 깔끔한 출력 위한 코드
pp = pprint.PrettyPrinter(indent=4)
print(pp.pprint(contents))
```

```python
### xml을 DataFrame으로 변환하기 ###
from os import name
import xml.etree.ElementTree as et
import pandas as pd
import bs4
from lxml import html
from urllib.parse import urlencode, quote_plus, unquote
```

```python
xml_obj = bs4.BeautifulSoup(contents,'lxml-xml')
rows = xml_obj.findAll('record')
print(rows)
```

```python
# 각 행의 컬럼, 이름, 값을 가지는 리스트 만들기
row_list = [] # 행값
name_list = [] # 열이름값
value_list = [] #데이터값

# xml 안의 데이터 수집
for i in range(0, len(rows)):
    columns = rows[i].find_all()
    #첫째 행 데이터 수집
    for j in range(0,len(columns)):
        if i ==0:
            # 컬럼 이름 값 저장
            name_list.append(columns[j].name)
        # 컬럼의 각 데이터 값 저장
        value_list.append(columns[j].text)
    # 각 행의 value값 전체 저장
    row_list.append(value_list)
    # 데이터 리스트 값 초기화
    value_list=[]
```
